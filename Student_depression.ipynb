#Student Depression Dataset

This dataset provides detailed information on students' mental health and associated factors. It is designed to help analyze patterns and potential predictors of depression among students. The data includes demographic details, academic and work-related pressures, lifestyle habits, and mental health indicators. Researchers can utilize this dataset to identify risk factors for depression and develop early intervention strategies, while ensuring ethical handling of sensitive personal data.

link to the source: https://www.kaggle.com/datasets/adilshamim8/student-depression-dataset

#Field Descriptions

- id: A unique identifier for each student in the dataset.

- Gender: The student’s gender
- Age: The student's age in years.

- City: The location or region where the student resides

- Profession: The field of study or work

- Academic Pressure: Indicates the level of academic stress the student experiences, such as pressure from exams, deadlines, or academic expectations.

- Work Pressure: Reflects the stress associated with employment responsibilities, relevant for students who work while studying.

- CGPA: The student's cumulative grade point average, representing their academic performance.

- Study Satisfaction: Measures how satisfied the student feels with their studies, which may influence their overall mental well-being.

- Job Satisfaction: Indicates the level of satisfaction with the student’s job or work environment, where applicable.

- Sleep Duration: The average number of hours the student sleeps per day—an essential factor in mental and physical health.

- Dietary Habits: Describes the student’s eating patterns and nutrition, which may impact their mood and well-being.

- Degree: The academic degree or program the student is currently pursuing.

- Have you ever had suicidal thoughts?:
A binary indicator (Yes/No) showing whether the student has ever experienced suicidal ideation.

- Work/Study Hours: The average number of hours per day the student spends on work or academic activities, which may affect stress levels.

- Financial Stress: Assesses the stress the student feels due to financial issues—an important aspect of mental health.

- Family History of Mental Illness: Indicates whether the student has a family history of mental illness (Yes/No), a potential risk factor for depression.

- Depression
The main outcome variable, indicating whether the student is currently experiencing depression (YES(1)/NO(0)).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.read_csv('student_depression_dataset.csv')
```

```python
df=pd.read_csv('student_depression_dataset.csv')
```

```python
df.info()
```

```python
df.shape
```

```python
df['Sleep Duration'] = df['Sleep Duration'].astype(str).str.replace("'", "").str.strip()

mapping = {
    'Less than 5 hours': 4,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9,
    'Unknown': np.nan
}


df['Sleep Duration'] = df['Sleep Duration'].map(mapping)

df['Sleep Duration'] = df['Sleep Duration'].astype(float)
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df['Have you ever had suicidal thoughts ?']=df['Have you ever had suicidal thoughts ?'].astype(str).str.strip().str.lower().map({'yes': True, 'no': False})
```

```python
df['Financial Stress'] = pd.to_numeric(df['Financial Stress'], errors='coerce')
```

```python
df['Family History of Mental Illness']=df['Family History of Mental Illness'].astype(str).str.strip().str.lower().map({'yes':True,'no':False})
```

```python
df['Degree'] = df['Degree'].str.strip("'")
```

```python
#Gender Distribution
ax=sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
for bar in ax.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,f'{int(height)}',
             ha='center', va='bottom')
plt.show()
```

```python
#Distribution of depression among students
ax=sns.countplot(data=df, x='Depression')
for bar in ax.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,f'{int(height)}',
             ha='center', va='bottom')
plt.title('Distribution of Depression Among Students')
plt.xticks(ticks=[0, 1], labels=['No Depression (0)', 'Depression (1)'])
plt.show()
```

```python
# Age Distribution by Gender and Depression
sns.boxplot(data=df, x='Gender', y='Age', hue='Depression')
plt.title('Age Distribution by Gender and Depression')
plt.show()
```

```python
#Top 10 cities by number of students along with their percentage.
df_city= df['City'].value_counts().head(10)
df_city_percent= (df['City'].value_counts()/df['City'].value_counts().sum()*100).round(2).astype(str)+'%'
result=pd.concat([df_city, df_city_percent], axis=1)
result.columns=['Count', 'Percentage']
result.sort_values('Count',ascending=False).head(10)
```

```python
#Top 10 cities with the highest number of students who have depression
depression_only = df[df['Depression'] == 1]
df_city_depression=depression_only['City'].value_counts().head(10)
df_city_depression_percent=(depression_only['City'].value_counts(normalize=True).head(10)*100).round(2).astype(str)+'%'
result=pd.concat([df_city_depression,df_city_depression_percent],axis=1)
result.columns=['Count','Percentage']
result
```

```python
#Comparison of Total Students vs Students with Depression by City

top_10_cities = df['City'].value_counts().head(10).index
df_city_all = df[df['City'].isin(top_10_cities)]['City'].value_counts()
depression_only = df[(df['City'].isin(top_10_cities)) & (df['Depression'] == 1)]
df_city_depression = depression_only['City'].value_counts()

result = pd.concat([df_city_all, df_city_depression], axis=1)
result.columns = ['All Students', 'With Depression']
result = result.fillna(0).astype(int)

result.plot(kind='bar', figsize=(10,6))
plt.title('Comparison of Total Students vs Students with Depression by City')
plt.ylabel('Number of Students')
plt.xlabel('City')
```

#Conclusion:
The dataset includes 27,901 students, with slightly more males (15,547) than females (12,354). Overall, 58.6% of students are experiencing depression, with males showing a higher depression rate than females. The average student age is 26, ranging from 18 to 59 years. Geographically, cities like Kalyan, Hyderabad, and Srinagar have the largest student populations and the highest depression cases, indicating areas for targeted mental health support.

Most students face moderate to high academic pressure, which correlates strongly with increased suicidal thoughts. At the highest pressure level, over three times more students report suicidal ideation compared to those who don’t, highlighting the need for academic and mental health support.

Work-related stress and job satisfaction are minimal for most students, aligning with the fact that the majority are likely focused on studies rather than employment. Study satisfaction varies, with many students moderately satisfied but fewer feeling very satisfied.

Academically, students perform reasonably well with an average CGPA of 7.7. However, dietary habits tend toward unhealthy or moderate, suggesting a need for improved nutrition awareness.

Financial stress affects many students, especially at moderate to high levels, and this is closely linked to higher rates of depression. Family history of mental illness is nearly evenly split among students, but those with such a history show a notably higher prevalence of depression.

Sleep patterns are varied, with many students sleeping only 4 to 7.5 hours. Shorter sleep durations strongly associate with higher depression rates and suicidal thoughts. Additionally, students studying or working long hours often report shorter sleep durations, increasing their risk of mental health problems.

Depression rates vary by academic program, with Class 12 students experiencing the highest number of cases. Across all levels, students dedicating more than 10 hours daily to work or study show significantly higher depression rates, underscoring the impact of prolonged academic and work demands on mental health.

#Recomendation:

- Support at-risk students: Focus mental health efforts on male students and those in cities with higher depression rates.

- Ease academic stress: Offer workshops and counseling to help students handle pressure better.

- Promote good sleep: Encourage students to get enough rest, especially those juggling long study or work hours.

- Help with money worries: Provide financial advice or aid to students struggling with financial stress.

- Check family history: Use family mental health info to spot students who might need extra care.

- Encourage healthy eating: Share simple tips to improve diet and boost well-being.

- Balance workload: Help students avoid too many study or work hours to reduce burnout.
